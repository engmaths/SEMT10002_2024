4. Compute the factorial of 10. Recall that the factorial of an integer n is defined as n! =
n × (n − 1) × (n − 2) × . . . 2 × 1.

````python
n = 10

for i in range(n-1, 0, -1):
    n = n*i
    
print(n)
````

3. Find the mistake(s) in the following program, which is meant to find the greatest power of 4 that is smaller than 200:	
```python
exponent = 0
power = 4 * exponent

while power > 400: 
    result = power
    power = 4 * exponent 

print('largest power = ', result) 
```



There are three mistakes here:

1. controlling expression after the `while` keyword should be `power < 200`, not `power > 200`
2. the value of `exponent` should be increased by 1 with each iteration of the loop 
3. `power` should be computed as `4**exponent`, not `4*exponent`

````python
exponent = 0
power = 4 ** exponent

while power < 200:
    
    result = power
    
    power = 4 ** exponent 
    
    exponent += 1
    
print('largest power = ', result) 
````

1. Determine how many positive cubic numbers are less than 2,000. Recall that cubic number is a number of the form $n^3$ where n is an integer.

````python
counter = 0

while counter**3 < 2000:
    stored_result = counter
    counter += 1
    
print(stored_result)
````

Note this could be achieved with a for loop but the code is longer

````python
num_cubes = 0

for n in range(1, 2000):
    
    if n**3 < 2000:
        num_cubes += 1
        
    else:
        break
        
print(num_cubes)
````

2. Write a program that calculates how many years it would take for the value of a savings
account to exceed £400, if the initial (and only) deposit made is £100 and the annual interest
is 5%. Answer = 29 years

4. The value of π can be approximated using the Leibniz formula:
$$ \pi_N = \sum_{n = 0}^{N} \frac{8}{(4n+1)(4n+3)} $$
where N is a large number. Taking the limit as N → ∞ produces the exact value of π,
but this requires evaluating an infinite number of terms, which is impossible on a computer.
Therefore, we can only approximate the value of π by using a finite number of terms in the
sum. Use this formula and nested loops, compute approximations to π by letting N equal
each value in the sequence N = [100, 1, 000, 10, 000].

````python
# create a variable for the approximate value of pi
pi_N = 0

# number of terms in the series
N = 10000

# use a for loop to compute the series
for n in range(N):
    pi_N += 8 / (4 * n + 1) / (4 * n + 3)
    
print(pi_N)
````

# Exercise 4 - Nested loops

1. Use nested loops to print a 5 x 5 multiplication table 

````python
# outer loop to iterate from 1 to 5
for i in range(1, 6):
    
    # nested loop to iterate from 1 to 5
    for j in range(1, 6):
        
        # print multiplication followed by tab space
        print(i * j, end='\t')
        
    # New line after inner loop completes
    print()
````

2. Use two for loops to compute the double sum
$$S = \sum_{i = 1}^{10} \sum_{j = 0}^{5} j^2(i + j)$$


````python
s = 0

for i in range(1,11):
    for j in range(0,6):
        s += j **2 * (i + j)

print(s)
````
